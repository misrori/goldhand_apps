## Import

```python
from goldhand import *
import warnings

pd.set_option('display.max_rows', 70)
warnings.filterwarnings('ignore')
```

### Tw class

```python
tw = Tw()
# stock data frame more than 8000 stock data from TradingView(https://www.tradingview.com/screener/)
display(tw.stock.head(2).T)

# Crypto data frame top 300 crypto data from TradingView(https://www.tradingview.com/crypto-coins-screener/)
display(tw.crypto.head(2).T)
```

```python
# Plot title
tw.get_plotly_title('TSLA')
```

```python
# Sector location of a ticker
tw.get_sec_plot('TSLA')
```

```python
# Industry location of a ticker
tw.get_ind_plot('COIN')
```

### GoldHand class
#### Get historical data and prepar for further analytics, also visualize it, with local minimums and maximums
#### It works with stock (for example: TSLA, AAPL) and crypto (for example: BTC-USD, ETH-USD)

```python
t = GoldHand('AAPL')
t.df.tail(1).T
```

```python
# display the local min and maximum values
t.df.loc[t.df['local_text']!=''].head(6)
```

```python
# with basic settings
ticker= 'BTC-USD'
t = GoldHand(ticker)
p = t.plotly_last_year(tw.get_plotly_title(ticker))
p.show()
```

```python
p = t.plotly_last_year(tw.get_plotly_title(ticker), ndays=1000)
p.show()
```

```python
p = t.plotly_last_year(tw.get_plotly_title(ticker), plot_height=900, ndays=500, ad_local_min_max=False)
p.show()
```

```python
# with base settings
ticker= 'ETH-USD'
t = GoldHand(ticker)
p = t.plot_goldhand_line(tw.get_plotly_title(ticker))
p.show()
```

```python
p = t.plot_goldhand_line(tw.get_plotly_title(ticker), ndays=1600)
p.show()
```

```python
p = t.plot_goldhand_line(tw.get_plotly_title(ticker), plot_height=900, ndays=800, ad_local_min_max=False)
p.show()
```
